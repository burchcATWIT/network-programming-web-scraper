```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

Starting with some utility methods. These are to save the html resonse from the webpage. A problem with webscraping is that bots we write can request web pages more times than we would like. These methods prevent that from happening.

```python
#from https://www.learndatasci.com/tutorials/ultimate-guide-web-scraping-w-python-requests-and-beautifulsoup/
def save_html(html, path):
    with open(path, 'wb') as file:
        file.write(html)


def open_html(path):
    with open(path, 'wb') as file:
        return file.read()
```

This is an http request and response done in our python script. The html response is saved using our methods from before

```python
URL = 'https://www.cardplayer.com/poker-players'
response = requests.get(URL)
save_html(response.content, 'pokersite')
```

Beatiful soup parser we imported to process the html response we recieved.

```python
soup = BeautifulSoup(response.content, 'html.parser')
rows = soup.select('tbody tr')
name = rows[0].select_one('td').text.split()
print(rows)
```

Processing the text in the columns of the html table

```python
cols = []
for row in rows:
    col = row.find_all('td')
    col = [atr.text.strip() for atr in col]
    cols.append(col)
print(cols)
```

Placing our data in a dataframe to view and manipulate

```python
index = ['Name', 'Country', 'Earnings', 'Cache']
poker_df = pd.DataFrame(cols, columns=index)
poker_df
```
